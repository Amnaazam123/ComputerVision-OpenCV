```python
import numpy as np
import math
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = plt.imread("nobita.jpg")
plt.imshow(img)
grayScale = cv.imread("nobita.jpg",0)
```

```python
plt.imshow(grayScale,cmap="gray")
```

```python

kernel=[
    [-1/3,0,1/3],
    [-1/3,0,1/3],
    [-1/3,0,1/3]
]
sum=0

fx = np.zeros((grayScale.shape[0], grayScale.shape[1]), dtype = int)
for i in range(1,grayScale.shape[0]-1):
    for j in range(1,grayScale.shape[1]-1): 
        sum=0
        x = i - 1
        for k in range(0,3):
            y = j - 1
            for l in range(0,3):
                sum = sum + (kernel[k][l] * grayScale[x][y])
                y = y + 1
            x = x + 1
        fx[i][j] = sum
plt.imshow(fx,cmap="gray")
```

```python
kernel=[
    [-1/3,-1/3,-1/3],
    [0,0,0],
    [1/3,1/3,1/3]
]
sum=0

fy = np.zeros((img.shape[0], img.shape[1]), dtype = int)
for i in range(1,img.shape[0]-1):
    for j in range(1,img.shape[1]-1):
        sum=0
        x = i - 1
        for k in range(0,3):
            y = j - 1
            for l in range(0,3):
                sum = sum + (kernel[k][l] * grayScale[x][y])
                y = y + 1
            x = x + 1
        fy[i][j] = sum

plt.imshow(fy,cmap="gray")
```

```python
edges = np.zeros((img.shape[0], img.shape[1]), dtype = int)
for i in range (0,img.shape[0]):
    for j in range (0,img.shape[1]):
        edges[i][j] = math.sqrt((fx[i][j]**2)+(fy[i][j]**2))
        
plt.imshow(edges,cmap="gray")
```
