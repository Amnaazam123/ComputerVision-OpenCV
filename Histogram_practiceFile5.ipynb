```python
fig=plt.figure(figsize=(10,5))     #10x5 ka aik khali box bna jis me ab wo images aein gi jo plt read krta gaya
                                   #images aen gi kitni, it depends upon number of subplots(if not subplots then it will adjust the whole readed image in figsize) 
fig.add_subplot(2,3,1)  #is box me 2 rows aur 3 cols(total max 6 subplots) hon gay aur us k pehle subplot me image add ho gi
plt.imshow(im)

fig.add_subplot(2,3,2)  #dosray subplot me image add ho gi    
plt.imshow(im)

fig.add_subplot(2,3,3)
plt.imshow(im)

fig.add_subplot(2,3,4)
plt.imshow(im)

fig.add_subplot(2,3,5)
plt.imshow(im)

fig.add_subplot(2,3,6)
plt.imshow(im)
```

```python
#rotate image
plt.imshow(np.rot90(im))
```

```python
plt.imshow(np.rot90(np.rot90(np.rot90(im))))
```

```python
red = im.copy()
plt.imshow(red)
```

```python
red[:,:,(1,2)]=0
plt.imshow(red);
```

```python
green = im.copy()
green[:,:,(0,2)]=0
plt.imshow(green)
```

```python
blue = im.copy()
blue[:,:,(0,1)]=0
plt.imshow(blue)
```

```python
fig=plt.figure(figsize=(10,10))

fig.add_subplot(1,3,1)
plt.imshow(red)

fig.add_subplot(1,3,2)
plt.imshow(green)

fig.add_subplot(1,3,3)
plt.imshow(blue)
```

```python
fig=plt.figure(figsize=(10,10))
images = np.concatenate((red,green,blue),axis=1)
plt.imshow(images)

```

```python
fig=plt.figure(figsize=(10,10))
images = np.concatenate((red,green,blue),axis=0)
plt.imshow(images)
```

```python
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.title('Original')
plt.imshow(im)

fig.add_subplot(1, 2, 2)
plt.title('Trimmed')
plt.imshow(im[400:600, 400:1000, :])         #triming image
```

```python
#below histogram shows there is a frequency of almost 180 of those colors ranging from 180 to 210 and so on. 
#To get value of frequency more precisely you can increase no of bins.
#since the graph is right skewed it means image is more brighter(low contrast image)

plt.hist(np.ravel(im),bins=10)          
plt.show()
```
