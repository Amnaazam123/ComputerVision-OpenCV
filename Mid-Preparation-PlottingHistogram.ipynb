```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("brighter.jpg",0)
```

```python
plt.imshow(img,cmap="gray")
```

```python
plt.hist(img,bins=10)
plt.xlabel("Gray-Level")
plt.ylabel("Count")
plt.show()
```

```python
img = cv.equalizeHist(img)
```

```python
plt.imshow(img,cmap="gray")
```

```python
plt.hist(img,bins = 10)
plt.show()
```
