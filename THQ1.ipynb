```python
import cv2 as cv
import matplotlib.pylab as plt
img = plt.imread("cat.jpg")
plt.imshow(img)
```

```python
import numpy as np
rows=img.shape[0]
cols=img.shape[1]
greyImg=np.zeros((rows,cols))
i=0
j=0
while i<rows:
    while j<cols:
        sum=int(img[i,j,0])+int(img[i,j,1])+int(img[i,j,2])
        sum=int(sum/3)
        greyImg[i,j]=sum
        j=j+1
    i=i+1
    j=0;
  
print(greyImg)
plt.imshow(greyImg,cmap='gray')
```
