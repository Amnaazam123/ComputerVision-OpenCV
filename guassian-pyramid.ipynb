```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("nobita.jpg",0)
```

```python
#downscaling image
def downScaleImage(img):
    r = int(img.shape[0]/2)
    c = int(img.shape[1]/2)

    a=-1
    b=-1
    res=np.zeros((r,c),dtype=int)
    for i in range(0,img.shape[0]-1):
        b=-1
        if(i%2==0):
            a=a+1
        for j in range(0,img.shape[1]-1):
            if(i%2==0):
                if(j%2==0):
                    b=b+1 
                    res[a,b]=img[i,j]     

    return res
```

```python
fig = plt.figure(figsize=(20,10))

plt.subplot(151)
plt.imshow(img,cmap="gray")

I = cv.GaussianBlur(img, (5,5), 1, 1)
plt.subplot(152)
plt.imshow(I,cmap="gray")

I = cv.GaussianBlur(img, (5,5), 2, 2)
plt.subplot(153)
plt.imshow(I,cmap="gray")

I = cv.GaussianBlur(img, (5,5), 4, 4)
plt.subplot(154)
plt.imshow(I,cmap="gray")

I = cv.GaussianBlur(img, (5,5), 8, 8)
plt.subplot(155)
plt.imshow(I,cmap="gray")
```

```python
level1 = np.uint8(downScaleImage(img))

fig1 = plt.figure(figsize=(20,10))
plt.subplot(151)
plt.imshow(level1,cmap="gray")

I = cv.GaussianBlur(level1, (5,5), 1, 1)
plt.subplot(152)
plt.imshow(I,cmap="gray")

I = cv.GaussianBlur(level1, (5,5), 2, 2)
plt.subplot(153)
plt.imshow(I,cmap="gray")

I = cv.GaussianBlur(level1, (5,5), 4, 4)
plt.subplot(154)
plt.imshow(I,cmap="gray")

I = cv.GaussianBlur(level1, (5,5), 8, 8)
plt.subplot(155)
plt.imshow(I,cmap="gray")
```
