```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("nobita.jpg",0)
```

```python
#downscaling image
def downScaleImage(img):
    r = int(img.shape[0]/2)
    c = int(img.shape[1]/2)

    a=-1
    b=-1
    res=np.zeros((r,c),dtype=int)
    for i in range(0,img.shape[0]-1):
        b=-1
        if(i%2==0):
            a=a+1
        for j in range(0,img.shape[1]-1):
            if(i%2==0):
                if(j%2==0):
                    b=b+1 
                    res[a,b]=img[i,j]     

    plt.imshow(res,cmap="gray")
    return res
```
